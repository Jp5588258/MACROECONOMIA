• As ETFs (Exchange Traded Fund) se tratam de fundos que são negociados na bolsa de valores. Dito isso, analisando as características de uma ETF, suas vantagens e desvantagens em relação a uma carteira de investimentos individuais pode-se concluir que: 

• A diversificação na ETF ocorre de maneira instantânea, pois se trata de um conjunto de ativos subjacentes. 

• Ocorre uma certa facilidade em atingir mercados amplos com grandes empresas, por exemplo, ou mais específicos em determinada área. 

• Há uma boa questão de acessibilidade proporcionada pelas ETFs, pois mesmo com a transparência através de informações públicas direcionadas aos investidores, ocorre uma facilidade de negociação, assim como em ações, podendo ser negociadas ao longo do dia. 

• Há, geralmente, custo através de taxas de administração dessas ETFs, além de outros associados à gestão do fundo. 

• Por se tratar de um investimento em uma bolsa de valores apresentam, certamente, riscos de mercado.

• As ETFs apresentam uma complexidade que exige uma boa análise por parte do investidor. Tal complexidade está associada   a fatores como: o intuito e as expectativas sobre a ETF, características dos ativos subjacentes que ditam o risco dessa ETF, os modelos e estratégias de atuação desse fundo, etc.

• as ETFs podem apresentar modelos de utilização dos dividendos de maneiras diferentes, podendo ser distribuídos normalmente entre os investidores, ou serem reinvestidos.

    Em suma, uma ETF pode disponibilizar facilidades com a acessibilidade e diversificação, porém, como todo investimento, os riscos e o modelo das ETFs devem ser bem analisados pelos investidores, vendo qual se adequa melhor ao seu perfil.
